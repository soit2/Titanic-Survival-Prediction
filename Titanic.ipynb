```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
#preprocess or clean the data
#step 1 
#identify the needed attributes --> survived,pclass.sex,age
df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis="columns",inplace=True)
df.head()
```

```python
#step-2
#convert sex attribute values into numerical values by using one hot encoding

dummies=pd.get_dummies(df.Sex)
dummies.head()
```

```python
#step-3
#convert the obtained true false from the dummies DATAFRAME into binary values using astype(int)
dummies=dummies.astype(int)
dummies.head()
```

```python
#step-4
#fill the null values in age column
df[["Age"]].head(10)
```

```python
#step-4
#fill the null values in age column
a=df.Age.mean()
df=df.fillna(a)
df.head(10)
```

```python
df.head()
```

```python
#concat the obtained Sex values column
```

```python
df=pd.concat([df,dummies],axis="columns")
df.head(10)
```

```python
X=df[["Pclass","Age","Fare","female","male"]]
X.head()
```

```python
X=X.drop(["female","male"],axis="columns")
X.head()
```

```python
X=pd.concat([X,dummies],axis="columns")
```
